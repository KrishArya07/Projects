Bagging:

Bagging is a technique that reduce variance and help prevent overfitting by creating multiple subset of the training data with replacement, training a model on each subset and then averaging their prediction or using majority of votes.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris=load_iris()
x=iris.data
y=iris.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=0)
rf_model=RandomForestClassifier(n_estimators=100,random_state=42)
rf_model.fit(x_train,y_train)
```

```python
y_pred=rf_model.predict(x_test)
acc=accuracy_score(y_test,y_pred)
acc
```

Boosting:

Boosting is a sequential ensemble technique that adjust the weight of
incorrectly classfied so that subsequent models focus more on difficult
cases. The final prediction is a weighted sum of the predictions from all
models .

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris=load_iris()
x=iris.data
y=iris.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
base_estimator=DecisionTreeClassifier(max_depth=1)
ada_model=AdaBoostClassifier(estimator=base_estimator,n_estimators=50,random_state=42)
ada_model.fit(x_train,y_train)
```

```python
y_pred=ada_model.predict(x_test)
y_pred
```

```python
y_test
```

```python
acc=accuracy_score(y_test,y_pred)
acc
```

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
iris=datasets.load_iris()
x=iris.data
y=iris.target
print(y)
x=x[y!=2]
y=y[y!=2]
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_train
```

```python
svm_model=SVC(kernel='rbf',gamma='scale',C=1.0,random_state=42)
svm_model.fit(x_train,y_train)
y_pred=svm_model.predict(x_test)
acc=accuracy_score(y_test,y_pred)
acc
```
